# Tags 

## Merge tables

```python
unsure_user = list(equipment["created_user"][equipment["created_user"] != equipment["updated_user"]]) 
equipment = equipment[~equipment["created_user"].isin(unsure_user)]
print("number of unsure user: " + str(len(unsure_user)))
```

```python
equipment_user = equipment[["id", "company_id", "created_user"]].merge(member[["id"]],
                                                     how="left",
                                                     left_on="created_user",
                                                     right_on="id",
                                                     suffixes=('_equipment', '_member'))

equipment_user.head()
```

```python
tag_eq_user = tag_eq[["equipment_id", "tag_id"]].merge(equipment_user,
                                         how="left",
                                         left_on="equipment_id", 
                                         right_on="id_equipment",  
                                        suffixes=(False, False))

tag_eq_user.head()
```

```python
tag_eq_user = tag_eq_user.merge(tag[["id", "type"]],
                 how="left",
                 left_on="tag_id",
                 right_on="id",
                 suffixes=(False, "_type"))

tag_eq_user = tag_eq_user.rename({
    columns="id_member": "member_id"
}).drop(columns="id_equipment")

tag_eq_user.head()
```

```python
tag_eq_user_company = (tag_eq_user.
                           merge(company[["id", "name"]],
                                 how="left",
                                 left_on="company_id",
                                 right_on="id",
                                 suffixes=("", "_company")).
                            drop(columns=["id_company", "id", "created_user"]).
                            rename(columns={"name": "company_name",
                                           "id_member": "member_id"}))
tag_eq_user_company                      
```

```python
tag_eq_user_company.to_csv("tag_eq_user_company.csv", index=False)
```

# Device count

## per member

```python
(equipment_user.
    groupby(["id_member"]).
    agg({'id_equipment': 'count'}).
    rename(columns={'id_equipment': 'n_device'}))
```

## per company 

```python
equipment_company = equipment[["id", "company_id", "created_user"]].merge(company[["id", "name"]],
                                                     how="left",
                                                     left_on="company_id",
                                                     right_on="id",
                                                     suffixes=('_equipment', '_company')).drop(columns = "id_company")
equipment_company.head()
```

```python
s1 = (equipment_company.
    groupby("name")['created_user'].
    nunique())
    
s2 = (equipment_company.
    groupby("name")['id_equipment'].
    nunique())

eq_user_company =  (pd.concat([s1, s2], axis=1).
                         rename(columns={'created_user': 'n_user',
                         'id_equipment': 'n_equipment'}).
                         assign(eq_user_ratio = lambda x: x.n_user / x.n_equipment))

eq_user_company.to_csv("eq_user_ratio.csv")
eq_user_company
```

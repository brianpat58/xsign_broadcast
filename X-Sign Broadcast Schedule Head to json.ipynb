```python
import pandas as pd
from pandas.io.json import json_normalize
import json
import time
```

```python
fileNM = "xs_schedule_head.csv"
schedule_head = pd.read_csv(fileNM)
```

```python
schedule_head.dtypes
```

```python
schedule_head.head()
```

```python
setData = json.loads(str(schedule_head['description'][0]))
schedule_head_df = json_normalize(setData)
```

```python
schedule_head_df
```

```python
# schedule_head_df = pd.concat([schedule_head_df, data], axis = 1, ignore_index=True, sort=False)

for i in range(1, len(schedule_head)):
    data = json_normalize(json.loads(str(schedule_head['description'][i])))
    schedule_head_df = pd.concat([schedule_head_df, data], axis = 0, ignore_index=True, sort=True)
```

```python
schedule_head_df.dtypes
```

```python
schedule_head_df.head()
```

```python
df2 = schedule_head['created_at']
df3 = schedule_head['company_name']
df4 = schedule_head['nation_name']
df5 = schedule_head['region_name']
result = pd.concat([schedule_head_df['for_tag'], df2, df3, df4, df5], axis = 1)
result_new = result.drop_duplicates()
```
